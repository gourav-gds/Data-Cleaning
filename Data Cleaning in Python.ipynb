```python
features = round((100 * books.isnull().sum() / len(books.index)), 2)
print(features.index)
```

```python
features.values
```

```python
#Extract all those features whose null value percentage is > 50%
list(features[features.values > 50].index)
```

```python
drop_features = list(features[features.values > 50].index)
books.drop(labels = drop_features, axis = 1, inplace = True)
books.info()
```
